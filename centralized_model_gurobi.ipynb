```python
filename = 'CASE23_baseline.xlsx'
param = Pnew(filename=filename,scid=0)
param.create_instance(sample=False)
graph = param.get_graph()
param.to_json()
params = json.load(open('data_1new.json'))

ratios = [[0,0],[1,0],[1,500],[1,5000]] #
#ratios = [[0,0]]
simulations = []
for ratio in ratios:
    file_segments = filename.split('/')
    xlsx_name = file_segments[-1][:-5].split('_')[-1]
    print(f'Scenario_{xlsx_name}_ratio_{ratio[0]}_{ratio[1]}')
    model = 0
    model = CentralizedSinglePeriod(params=params,time_neutral=False,ratios=ratio,unmet_penalty=[1],relax=False)
    print('GurobipyResults')
    start_time = time.time()
    model.solve(silent=False)
    end_time = time.time()
    print(end_time-start_time)
    print('------')
    model.solution_visualization()
    #Running simulation
    disrupted_agents = {'All':['log_norm',0.3]}
    df_sim_res,scenarios = model.generate_stochastic_leadtime(n_samples=300,disrupted_agents=disrupted_agents)
    agg_stats = pd.DataFrame(df_sim_res.groupby(by=['i','j','k'])['lateness'].mean())
    agg_stats.columns = ['mean']
    agg_stats['std'] = pd.DataFrame(df_sim_res.groupby(by=['i','j','k'])['lateness'].std())['lateness']
    df_sim_res['scenario'] = f'{ratio[0]}:{ratio[1]}'
    simulations.append(df_sim_res)
```

```python
k_data = model.condensed_statistics_compare(simulations,['#31C7DE','#8ECDE8','#42A1FF','#140694'],'Veh2_S7_simulation')
```

```python
stats = pd.concat(simulations, ignore_index=True)
stats['Type'] = stats['j'].map(model.m._V_type)
hist_data = stats[stats['Type']=='Retail']
hist_data = hist_data[hist_data['k']=='Vehicle2']
print(filename)
for scenario in hist_data['scenario'].unique():
    print(scenario)
    scenario_hist = hist_data[hist_data['scenario']==scenario]
    arrays = []
    for r in scenario_hist.index:
        arrays.append(np.repeat(a=scenario_hist.loc[r,'lateness'],repeats=int(scenario_hist.loc[r,'lateness'])))
    frec_arr = np.concatenate(arrays)
    print('mean:',frec_arr.mean())
    print('std:',frec_arr.std())
```

```python
#f,a = plt.subplots(3,1)
{'Dist': 'wheat', 'Manuf': 'lightskyblue', 'Part': 'lightsteelblue', 'Retail': 'mistyrose'}
agg_stats = results[['Gen Mean','Gen Std']]
agg_stats = agg_stats.sort_index(ascending=False)
from matplotlib.transforms import Affine2D

fig, ax = plt.subplots(figsize=(10,6))
displacement = [-0.1,0,0.1]
color = ['#607FAB','#84AFDB','#DEC188']
style = ['o','^','s']

for i,disrpt_agt in enumerate(list(agg_stats.index.levels[0])[::-1]):
    print(disrpt_agt)
    ind_agg_stats  = agg_stats.xs(disrpt_agt).reset_index()
    trans = Affine2D().translate(displacement[i],0.0)+ax.transData
    ax.errorbar(ind_agg_stats['Configuration'], ind_agg_stats['Gen Mean'], ind_agg_stats['Gen Std'], linestyle='None', marker=style[i],markersize=15,capsize=8,transform=trans,label=f'{disrpt_agt}',linewidth=5,color=color[i])
plt.legend(bbox_to_anchor=(1, 1.2),ncol=3,borderaxespad=0,title='Disrupted entity')
plt.ylabel('Days of lateness in delivery')
plt.xlabel('SC network topology')
plt.savefig('TopologyAveragePerformance.pdf',bbox_inches='tight')
```
